```python
import pandas as pd
df = pd.read_csv('avia.csv')
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df
```

```python
#сдесь приводим дату/время к нужному формату
df['scheduled_departure'] = pd.to_datetime(df['scheduled_departure'], format='%Y-%m-%dT%H:%M:%SZ')
```

```python
#сдесь цены на топливо по зимним месяцам 
df.loc[df.scheduled_departure.dt.month == 1, 'fuel_price (kg)'] = 41.435
df.loc[df.scheduled_departure.dt.month == 2, 'fuel_price (kg)'] = 39.553
df.loc[df.scheduled_departure.dt.month == 12, 'fuel_price (kg)'] = 47.101
```

```python
df.model.unique()
```

```python
# сдесь прописан расход топлива для используемых самолетов на рейсах в белгород и в москву
df.loc[df.model == 'Sukhoi Superjet-100', 'fuel_per_hour'] = 1700
df.loc[df.model == 'Boeing 737-300', 'fuel_per_hour'] = 2600
```

```python
# сдесь простое уравнение для нахождения потерь/затрат на рейсе
df['lost'] = df['minutes'] / 60 * df['fuel_per_hour'] * df['fuel_price (kg)']
```

```python
df['cash'] = df['rubles'] - df['lost']
```
